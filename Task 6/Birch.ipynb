## Selecting the features

```python
X = mall[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
X.head()
```

## Finding the suitable branching factor

```python
silhouette_list = []
for i in range(2, 20):
    birch = Birch(branching_factor=i, n_clusters=5, threshold=3.5)
    birch.fit(X)
    birch_labels = birch.labels_
    score = metrics.silhouette_score(X, birch_labels)
    silhouette_list.append(score)
    
```

```python
plt.figure(figsize = (10,6))
plt.plot(list(range(2, 20)), silhouette_list)
plt.xlabel("min_samples")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
## Here we have high silhouette score for the min_samples being somewhere between 7 to 7.5
```

## Training the model

```python
birch = Birch(branching_factor=15, n_clusters=5, threshold=2.5)
```
